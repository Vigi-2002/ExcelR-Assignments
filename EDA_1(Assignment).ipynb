## Data Cleaning and Preparation

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Cardiotocographic.csv')
```

```python
data.head()
```

```python
data.info()
```

### Checking Duplicate values

```python
data[data.duplicated()]
```

```python
#Dropping duplicate values
data.drop_duplicates(inplace = True)
```

### Checking for Null values

```python
data.isna().sum()
```

### Filling Null values with mean of particular column

```python
LB_mean = np.round(data['LB'].mean(),1)
AC_mean = np.round(data['AC'].mean(),1)
DS_mean = np.round(data['DS'].mean(),1)
DP_mean = np.round(data['DP'].mean(),1)
MLTV_mean = np.round(data['MLTV'].mean(),1)
Width_mean = np.round(data['Width'].mean(),1)
Tendency_mean = np.round(data['Tendency'].mean(),1)
NSP_mean = np.round(data['NSP'].mean(),1)
```

```python
data.fillna({'LB' : LB_mean,
            'AC' : AC_mean,
            'DS' : DS_mean,
            'DP' : DP_mean,
            'MLTV' : MLTV_mean,
            'Width' : Width_mean,
            'Tendency' : Tendency_mean,
            'NSP' : NSP_mean}, inplace = True)
```

### Dropping Null Values

```python
data.dropna(inplace = True)
```

```python
data.info()
```

## Statistical Summary

### Calculate mean, median and standard deviation of each column

```python
for col in data.columns:
    mean = data[col].mean()
    median = data[col].median()
    standard_deviation = data[col].std()
    print(f'{col} : mean = {mean}, median = {median}, standard_deviation = {standard_deviation}')
```

### Interquartile Range(Identifying outliers)

```python
for col in data.columns:
    Q1 = data[col].quantile(0.25) #First Quartile (25th percentile)
    Q2 = data[col].quantile(0.75) #Second Quartile (75th percentile)
    IQR = Q2 - Q1 #Interquartile Range

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q2 + 1.5 * IQR

    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]
    print(f"Column: {col}")
    print(f"Q1: {Q1}, Q2: {Q2}, IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Outliers:\n{outliers}\n")
```

## Data Visualization

### Histogram

### Using histogram to visualize the distributions of various numerical variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in data.columns:
    plt.figure(figsize = (6,10))
    sns.histplot(data[col])
    plt.show()
```

## Barchart

```python
for col in data.columns:
    plt.figure(figsize = (20, 20))
    sns.countplot(x = col, data = data, palette = 'Set2')
    plt.show()
```
