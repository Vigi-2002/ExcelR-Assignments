## Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(scaled_data)
```

```python
wcss = np.round(model.inertia_,2) # Within cluster sum of squares
wcss
```

```python
model.labels_
```

```python
df['label'] = model.labels_ # Displays the cluster the particular point belongs to
df
```

```python
grp = df.groupby('label') # Grouping based on cluster(mean)
grp.agg('mean')
```

```python
for x, y in grp:
    print(y)
```

## Evaluation of K-Means algorithm

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(scaled_data, df['label'])
print(f'Silhouette Score: {sil_score:.2f}')
```

```python
wcss = []

for i in range(1,11): # Displays the winthin cluster sum of squares for '10' clusters
    model = KMeans(n_clusters=i)
    model.fit(scaled_data)
    inertia = np.round(model.inertia_,2)
    wcss.append(inertia)
wcss
```

## Elbow plot

```python
plt.figure(figsize=(6,3))
plt.title('Hyperparameter Tuning using Elbow Plot')
plt.xlabel('No of Clusters (k)')
plt.ylabel('WCSS')
plt.plot(range(1,11), wcss,
         marker = 'H', markerfacecolor='orange',
         linestyle='dotted'
        )
plt.show()
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3,linkage='complete', metric='euclidean')
model.fit(scaled_data)
```

```python
model.labels_
```

```python
df['Labels'] = model.labels_
df
```

```python
grp = df.groupby('Labels') # Grouping based on cluster
grp.agg('mean')
```

```python
for x,y in grp:
    print(y)
```

## Linkage criteria

```python
from scipy.cluster import hierarchy
```

```python
plt.figure(figsize=(100, 100))
hr = hierarchy.linkage(scaled_data, method='complete')
dn = hierarchy.dendrogram(hr,color_threshold=1.25)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=0.5, min_samples=12)
```

```python
model.fit(scaled_data)
```

```python
model.labels_
```

```python
df['Labels'] = model.labels_
df
```

```python
grp= df.groupby('Labels') # Grouping based on clusters
grp.agg('mean')
```

```python
for x,y in grp:
    print(y)
```

## Evaluation (minPts)

```python
# Compute k-distances

from sklearn.neighbors import NearestNeighbors

minPts = 5
neighbors = NearestNeighbors(n_neighbors = minPts)
neighbors_fit = neighbors.fit(scaled_data)
distances, indices = neighbors_fit.kneighbors(scaled_data)
```

```python
k_distances = np.sort(distances[:, -1])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(k_distances)
plt.title(f'k-Distance Graph (minPts = {minPts})')
plt.xlabel('Points sorted by distance')
plt.ylabel(f'Distance to {minPts}th nearest neighbor')
plt.show()
```

```python
sil_score = silhouette_score(scaled_data, df['label'])
print(f'Silhouette Score: {sil_score:.2f}')
```

# Visualizing the clustering results 

```python
from sklearn.datasets import make_blobs
```

```python
kmeans = KMeans(n_clusters=4)
kmeans_labels = kmeans.fit_predict(scaled_data)

dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

agg_clustering = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_clustering.fit_predict(scaled_data)
```

```python
plt.figure(figsize=(50, 50))

# Plot K-Means clusters
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.5, label='K-Means', marker='o')

# Plot DBSCAN clusters (with special color for outliers)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=dbscan_labels, cmap='coolwarm', alpha=0.5, label='DBSCAN', marker='^')

# Plot Agglomerative Clustering clusters
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=agg_labels, cmap='plasma', alpha=0.5, label='Agglomerative', marker='s')
```
