## Standardizing the dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

## Visualing the features

```python
for col in df:
    plt.figure(figsize = (20, 20))
    sns.histplot(x = col, data = df)
    plt.title(f'Category Distribution for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Implementing PCA on the dataset

```python
from sklearn.decomposition import PCA
```

```python
# Considering all components

pca = PCA(n_components = 14)
pcomponents = pca.fit_transform(scaled_df)
pcomponents
```

```python
ratio = pca.explained_variance_ratio_
ratio
```

```python
vars = np.round(np.cumsum(ratio)*100,1) # Returns cumulative sum of varience
vars
```

```python
df.columns
```

```python
df[['Type', 'Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins'
    #'Color', 'Hue','Dilution', 'Proline'
   ]]
```

```python
pca = PCA(n_components = 10) # Components == features
pcomponents = pca.fit_transform(scaled_df)
pcomponents
```

## Building model using scaled dataset

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(scaled_df)
```

```python
wcss = np.round(model.inertia_,2)
wcss
```

```python
model.labels_
```

```python
df['label'] = model.labels_ # Grouping into clusters
df
```

```python
grp = df.groupby('label')
grp.agg('mean')
```

```python
for x, y in grp:
    print(y)
```

## Model Evaluation

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(scaled_df, df['label'])
print(f'Silhouette Score: {sil_score:.2f}')
```

```python
wcss = []

for i in range(1,11): # Displays the within clusters sum of squares for '10' clusters
    model = KMeans(n_clusters=i)
    model.fit(scaled_df)
    inertia = np.round(model.inertia_,2)
    wcss.append(inertia)
wcss
```

```python
plt.figure(figsize=(6,3))
plt.title('Hyperparameter Tuning using Scaled Dataset')
plt.xlabel('No of Clusters (k)')
plt.ylabel('WCSS')
plt.plot(range(1,11), wcss,
         marker = 'H', markerfacecolor='orange',
         linestyle='dotted'
        )
plt.show()
```

## Building model using PCA dataset

```python
model_pca = KMeans(n_clusters=3)
model_pca.fit(pcomponents)
```

```python
wcss = np.round(model_pca.inertia_,2)
wcss
```

```python
model_pca.labels_
```

```python
df['label'] = model_pca.labels_
df
```

```python
grp = df.groupby('label')
grp.agg('mean')
```

```python
for x, y in grp:
    print(y)
```
