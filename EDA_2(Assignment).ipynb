# Applying scaling techniques to numerical features

### Select numeric columns

```python
numeric_cols = df.select_dtypes(include = ['int64','float']).columns
df_numeric = df[numeric_cols]
```

### Applying Standard Scaling (Z-score normalization)

```python
from sklearn.preprocessing import StandardScaler

scaler_standard = StandardScaler()
df_standard_scaled = pd.DataFrame(scaler_standard.fit_transform(df_numeric), columns = numeric_cols)
df_standard_scaled
```

Standard scaling maintains the shape of the distribution, which is useful for models relying on standard deviation and covariance.

This scales the data so it has a mean of 0 and standard deviation of 1.

Best for: Linear Regression, Logistic Regression, Support Vector Machines (SVM), etc.

### Apply Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
df_minmax_scaled = pd.DataFrame(scaler_minmax.fit_transform(df_numeric), columns = numeric_cols)
df_minmax_scaled
```

Min-max scaling ensures that all features contribute equally to the model. It's especially helpful when features have different units or very different scales.


This scales the data to a range between 0 and 1.

Best for: Deep learning models, K-Means, etc.

## Encoding Techniques

### Identify categorical columns

```python
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols
```

### Applying One-Hot Encoding

### Applying One-Hot Encoding for cols. that have less than 5 categories

```python
one_hot_cols = [col for col in categorical_cols if df[col].nunique() <= 5]
one_hot_cols
```

```python
df_one_hot = pd.get_dummies(df[one_hot_cols], drop_first=True)
df_one_hot
```

Pros:

No ordinal relationship introduced: Each category is represented as a separate binary column.

Works well with most ML algorithms: Especially tree-based and linear models, which treat features independently.

Safe for nominal data: Best suited when categories are just labels (e.g., Male, Female, Other).



Cons:

High dimensionality: Creates many columns if a categorical feature has many unique values â†’ leads to sparse matrices.

Memory & computational cost: Large datasets with many categories may require more memory and slow down model training.

Can lead to the â€œcurse of dimensionalityâ€: More features = risk of overfitting, especially with limited data.

### Applying Label Encoding

### Applying Label Encoding for cols. that have more than 5 categories

```python
label_encode_cols = [col for col in categorical_cols if df[col].nunique() > 5]
label_encode_cols
```

```python
from sklearn.preprocessing import LabelEncoder

df_label_encoded = df[label_encode_cols].apply(LabelEncoder().fit_transform)
df_label_encoded
```

Pros:

Very efficient: Transforms categories into integers â†’ minimal memory usage.

Fast to implement: Just one line using LabelEncoder().

Good for tree-based models: Decision Trees and Random Forests donâ€™t rely on value magnitude, so labels can still be interpreted meaningfully.

Cons:

Imposes an ordinal relationship: red = 0, green = 1, blue = 2 makes it seem like blue > green > red, which may not be true.

Can confuse linear models: Algorithms like Linear Regression or SVM might interpret the encoded integers as ordered, which leads to misleading results.

## Feature Engineering

### Feature 1: Age Group

Rationale:

Age does not have a direct impact on income and job type. 

Grouping into broader categories helps models capture age-related trends without overfitting to specific numbers.

```python
df['age_group'] = pd.cut(df['age'],
                         bins=[0, 25, 45, 65, 100],
                         labels=['Young', 'Middle_Aged', 'Senior', 'Elderly'])

df['age_group']
```

### Feature 2: Net Capital

Formula: capital gain - capital loss

Rationale:

Combines both financial gain and loss into one meaningful measure of financial status or investment return, which may be more predictive of income level.

```python
df['net_capital'] = df['capital_gain'] - df['capital_loss']
```

```python
df['net_capital']
```

### Apply a transformation

```python
# Check skewness of numerical features

skewness = df[numeric_cols].skew().sort_values(ascending=False)
skewness
```

```python
# Choose a highly skewed feature

skewed_feature = 'capital_gain'
```

```python
# Apply log(1 + x) transformation

df[f'{skewed_feature}_log'] = np.log1p(df[skewed_feature])
```

```python
# Plot original

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df[skewed_feature], kde=True, bins=30, color='blue')
plt.title(f'Original Distribution: {skewed_feature}')
plt.tight_layout()
plt.show()
```

```python
# Plot log-transformed feature

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
sns.histplot(df[f'{skewed_feature}_log'], kde=True, bins=30, color='green')
plt.title(f'Log-Transformed: {skewed_feature}_log')
plt.tight_layout()
plt.show()
```

Original Skewness: 11.95 â†’ Highly right-skewed

After Log Transformation: 3.10 â†’ Much less skewed, better for modeling

Justification:

Skewed features can distort relationships in models like linear regression or neural networks.

Log transformations compress large values, making the data more normally distributed and model-friendly.

## Feature Selection

### Use the Isolation Forest algorithm (Identify and Remove outliers)

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.02, random_state=42)
outliers = iso_forest.fit_predict(df_numeric)
```

```python
# Add outlier flag to the original DataFrame

df['outlier'] = outliers
```

```python
# Keep only non-outliers (label = 1)

df_no_outliers = df[df['outlier'] == 1].drop(columns='outlier')
```

```python
# Show shape before and after removing outliers

print("Original shape:", df.shape)
print("After removing outliers:", df_no_outliers.shape)
```

Negative Effects of Outliers on Model Performance:

Distort Model Training: Outliers can bias coefficients in linear models or affect tree splits.

Reduced Accuracy: Models may try to fit noise instead of real patterns â†’ overfitting.

Skewed Metrics: Performance metrics (like RMSE or accuracy) can be dominated by a few extreme values.

Slower Convergence: Algorithms like gradient descent can struggle when extreme values dominate gradients.

### Apply PPS (find and discuss the relationships between features)

```python
import ppscore as pps
# Compute PPS matrix

pps_matrix = pps.matrix(df)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```

```python
# Plot PPS heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pps_matrix, cmap="viridis", annot=False)
plt.title("Predictive Power Score Matrix")
plt.show()
```

```python
# Compute Pearson correlation matrix (for numeric features only)
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Pearson Correlation Matrix")
plt.show()
```
