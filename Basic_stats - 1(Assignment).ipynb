### Identify outliers using IQR(Interquartile Range)

```python
for col in sales.select_dtypes(include=['number']).columns:
    Q1 = sales[col].quantile(0.25)  # First quartile (25th percentile)
    Q3 = sales[col].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = sales[(sales[col] < lower_bound) | (sales[col] > upper_bound)][col]
    print(f"Column: {col}")
    print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    print(f"Outliers:\n{outliers}\n")
```

## Barchart Analysis

### Identify categorical columns

```python
categorical_cols = sales.select_dtypes(include=['object', 'category']).columns
print(f"Categorical Columns: {categorical_cols}")
```

### Barchart

```python
for col in categorical_cols:
    plt.figure(figsize = (20, 20))
    sns.countplot(x = col, data = sales, palette = 'Set2')
    plt.title(f'Category Distribution for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Standardization of Numerical Variables

### Standardization numerical columns helps improve uniformity and ensures better model performance using Z-Score Normalization.

### Before standardization

```python
for col in numerical_cols:
    plt.figure(figsize = (20, 20))
    sns.histplot(x = col, data = sales, palette = 'Set2')
    plt.title(f'Category Distribution for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

### After standardization

```python
from sklearn.preprocessing import StandardScaler
numeric_df = sales.select_dtypes(include=['number'])

scaler = StandardScaler()

# Apply standardization to numerical columns
df_scaled = pd.DataFrame(scaler.fit_transform(numeric_df), columns = numeric_df.columns)

# Display standardized DataFrame
print("\nStandardized Data (Using StandardScaler):\n", df_scaled)
```

```python
for col in numeric_df:
    plt.figure(figsize = (20, 20))
    sns.histplot(x = col, data = sales, palette = 'Set2')
    plt.title(f'Category Distribution for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
